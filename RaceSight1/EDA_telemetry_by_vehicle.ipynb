```python
import pandas as pd 

# Load CSV file 

df = pd.read_csv('R2_barber_telemetry_data.csv')

#Excel File is very large, split into multiple files based on vehicle_number can make processing easier. 
for vehicle_number, group in df.groupby("vehicle_number"):
    filename = f"vehicle_{vehicle_number}.xlsx"
    group.to_excel(filename, index=False)
    print (f"Saved: {filename}")
```
